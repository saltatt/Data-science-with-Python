```python
df= pd.read_csv('/content/players_20.csv')
df.set_index('short_name',inplace=True)
df=df[['long_name','age','dob','height_cm','weight_kg','nationality','club','dob','weight_kg','overall']]
```

```python
#13
df = df.loc[:, ~df.columns.duplicated()]
df.columns = [''.join(col) if isinstance(col, tuple) else col for col in df.columns]
df['dob'] = pd.to_datetime(df['dob'])
filtered_index = df.query("age < 20 and overall > 80").index
df.loc[filtered_index, ['club', 'nationality']]
```

```python
#14
df = df.copy()
def extract_scalar(x):
    if isinstance(x, pd.Series):
        return x.iloc[0]
    return x

df['weight_kg'] = df['weight_kg'].apply(extract_scalar)
df['height_cm'] = df['height_cm'].apply(extract_scalar)

df['bmi'] = df['weight_kg'] / ((df['height_cm'] / 100) ** 2)
df['bmi'].sample(5)
```
