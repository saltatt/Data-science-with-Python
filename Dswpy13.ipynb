```python
df_series=df_netflix[df_netflix['type']=='TV Show'].copy()
df_series=df_series.assign(
    seasons=df_series['duration'].str.split(expand=True)[0].astype(int)
    )
```

```python
'''Для той же подвыборки сериалов постройте боксплот для seasons (горизонтальный, с
цветом &#39;green&#39;). Посчитайте Q1, Q3, IQR и найдите выбросы по правилу 1.5 * IQR.
Сколько таких выбросов? Выведите их названия (title).'''
df_series['seasons'].plot(kind='box',vert=False,color='green')
plt.show()
Q1=df_series['seasons'].quantile(0.25)
Q3=df_series['seasons'].quantile(0.75)
IQR =Q3-Q1
lower_bound = Q1-1.5*IQR
upper_bound = Q3+1.5*IQR
outliers = df_series[(df_series['seasons'] < lower_bound) | (df_series['seasons']>upper_bound)]
print(f'\nOutliers by rule IQR:{len(outliers)} from {len(df_series)}')
outliers
```

```python
'''Постройте барплот для колонки type (всего датасета). Есть ли редкие категории,
которые могут быть выбросами или ошибками? Затем проверьте колонку listed_in —
посчитайте топ-10 самых частых значений и постройте барплот для них.'''
df_netflix['type'].value_counts().plot(kind='bar')
plt.show()
df_netflix['listed_in'].value_counts().head(10).plot(kind='bar')
plt.show()
```

```python
'''8. В колонке country приведите все значения к нижнему регистру и удалите
ведущие/конечные пробелы. Выведите уникальные страны до и после очистки —
сколько уникальных стало меньше?'''
before = df_netflix['country'].nunique()
df_netflix['country'] = df_netflix['country'].str.lower().str.strip()
after = df_netflix['country'].nunique()
print("Before:",before)
print("After:",after)
print("Reduced by:", before - after)
```
