```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/laptop_price.csv',encoding='cp1252')
```

```python
pd.set_option('display.max_rows',5)
```

```python
# selected companies without HP
df[df['Company'] != 'HP']
```

```python
# selecting prices above 2000
df[df['Price_euros'] > 2000]
```

```python
#selecting screen sizes less 13
df[df['Inches'] < 13]
```

```python
#selecting notebooks
df[df['TypeName'] == 'Notebook']
```

```python
df[df['Company']=='Acer']
```

```python
#dividing  to 2 parts col inches
np.where(df['Inches']>15,'Big','Small')
```

```python
#adding it to df
df['Screen_size'] = np.where(df['Inches']>15,'Big','Small')
```

```python
df.head(5)
```

```python
#checking
df['Screen_size'].value_counts()
```

```python
#filtering
df['Price_categories']=np.where(df['Price_euros'] >1000,'Expensive','Cheap')
```

```python
#checking
df['Price_categories'].value_counts()
```

```python
#
cond = [
    (df['Inches']>16),
    (df['Inches']>14) &(df['Inches']<=16),
    (df['Inches']>12) & (df['Inches']<=14),
    (df['Inches']<=12)
]
ch = ['Too Big','Big','Small','Too Small']
df['Screen_size'] = np.select(cond,ch,default=' ')
```

```python
df['Screen_size'].value_counts()
```

```python
conditions = [
    (df['Price_euros']>3000),
    (df['Price_euros']>2000) & (df['Price_euros']<=3000),
    (df['Price_euros']>800) & (df['Price_euros'] <=2000),
    (df['Price_euros']<=800),
]
choices = ['Too Expensive','Expensive','Affordable','Cheap']
df['Price_tier'] = np.select(conditions,choices, default='')
```

```python
df['Price_tier'].value_counts()
```
